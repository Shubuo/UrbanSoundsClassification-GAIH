## Spectogram oluşturma
Ses dosyalarının spectrogramlarını oluşturun.
Create_spectrogram fonksiyonu, okuduğunuz ses dosyasını gönderdiğinizde size hazırlanmış spectrogramı dönecektir.


```python
# import the libraries
import matplotlib.pyplot as plt
import numpy as np
import librosa
import librosa.display
from scipy.io.wavfile import read
```

```python
def create_spectogram(y):
  spec = librosa.feature.melspectrogram(y=y)
  spec_conv = librosa.amplitude_to_db(spec,ref=np.max)
  return spec_conv
```

```python
path = ('/content/7061-6-0-0.wav')

#for i in path:
#path = np.array(path)
y, sr = librosa.load(path)
#create_spectogram(y)
plt.figure(figsize=(10,2))
D = create_spectogram(y)
img = librosa.display.specshow(D, y_axis='linear', x_axis='time', sr=sr)
''' ax[0].set(title='Linear-frequency power spectrogram')
ax[0].label_outer() 
librosa.output.write_wav('ses_kayit.wav', x, sr)
'''
```
